```python
import os
import pandas as pd
import tabula

base = '/content/drive/MyDrive/_599/Capstone/'
os.chdir(base)
```

# Sun Tables
https://gml.noaa.gov/grad/solcalc/

```python
sun_tables_path = '/content/drive/MyDrive/_599/Capstone/Data/Sunset Data/'
files = os.listdir(sun_tables_path)
files
```

```python
# header column names
header = ['day','jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']

for file_ in files:
  file_path = sun_tables_path + file_
  # tabula returns a list, 1st item is our data
  df = tabula.read_pdf(file_path, pandas_options={'header':None})[0]
  # write column names
  df.columns = header

  year_str = file_[:4]

  df.to_csv(sun_tables_path + f'sunset_{year_str}.csv',index=False)
```

# Reading In Sunset Files

```python
sunset_files = ['sunset_2018.csv',
                'sunset_2019.csv',
                'sunset_2020.csv',
                'sunset_2021.csv',
                'sunset_2022.csv']
sun_tables_path = '/content/drive/MyDrive/_599/Capstone/Data/Sunset Data/'

sunset_dict = {}
for file_ in sunset_files:
  file_path = sun_tables_path + file_
  # tabula returns a list, 1st item is our data
  df = pd.read_csv(file_path)

  # grab year from file, 1st removing .csv (:-4) following w/ removal of sunset (7:)
  year_str = file_[:-4][7:]

  sunset_dict[int(year_str)] = df


sunset_dict[2018]
```
