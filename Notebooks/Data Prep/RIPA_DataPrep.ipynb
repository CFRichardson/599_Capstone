```python
def null_ratios(df):
    nulls_ = (df.isnull().sum() * 100 / len(df)).to_frame().reset_index()
    nulls_.columns = ['feature', '%_missing']
    return nulls_.sort_values('feature')
```

# Feature Engineering

## Time Period

```python
def filter_func(x):
    hour = int(x['time_stop'][:2])
    if (hour < 12):
        return 0
    elif (hour >= 12 and hour <= 18):
        return 1
    else :
        return 2
        
ripa_final['time_period'] = ripa_final.apply(lambda row: filter_func(row), axis=1)
ripa_final.head()
```

## Sunset periods

```python
# list of hours between midnight and 6 am
early_morning_hours = list(range(0,7,1))
light = []

for idx, row in ripa_final.iterrows():
    date_list = row['date_stop'].split('-')
    year = int(date_list[0])
    month = int(date_list[1])
    day = int(date_list[2])

    sunset_time = sunset_dict[year].iloc[day-1, month] # day-1 due to index starting at 0
    sunset_hour = int(sunset_time[:2])
    sunset_minute = int(sunset_time[3:])
    
    # if sunset minute is past @ or past half hour, increase hour by 1
    if sunset_minute > 29:
        sunset_hour += 1

    stop_hour = int(row['time_stop'][:2])

    # if stop hour is b4 sunset_hour and not between the hours of 00 - 06am
    if (stop_hour < sunset_hour) and (stop_hour not in early_morning_hours):
        light.append(1)
    else:
        light.append(0)

ripa_final['light'] = light
```

### Month, Day of Week, & Hour Stopped

```python
ripa_final['month'] = ripa_final['date_stop'].apply(lambda row: int(row[5:7]))
ripa_final['day_of_week'] = ripa_final['date_stop'].astype('datetime64').dt.day_of_week
ripa_final['hour_stopped'] = ripa_final['time_stop'].apply(lambda row: int(row[:2]))
ripa_final.sample(5)
```

# Weather's Temperature Clean Up
Weather has 3 features tmin, tmean, & tmax.  Each record only needs one of the three dependent on the time of the day, giving the record an approximent temperature.

In example,

If a stop was done before sunrise, there is no light and thus the temperature is most likely at or near tmin.  If a stop was done after sunrise, but during the evening portion of the day temp will be most likely be tmean due to the fact that sunset is near and thus temperatures are slowly decreasing from tmax to tmean.

```python
temp = []
for idx, row in ripa_final.iterrows():
    light = row['light']
    time_period = row['time_period']

    if light == 0:
        if time_period == 0:
            temp.append(row['tmin'])
        elif time_period == 1:
            temp.append(row['tmean'])
        elif time_period == 2:
            temp.append(row['tmin'])

    elif light == 1:
        if time_period == 0:
            temp.append(row['tmean'])
        elif time_period == 1:
            temp.append(row['tmax'])
        elif time_period == 2:
            temp.append(row['tmean'])

ripa_final['temp'] = temp
ripa_final = ripa_final.drop(columns=['tmin', 'tmean', 'tmax'])
ripa_final.head()
```

```python
# several records were duplicated during the merging processees
print('DF Shape Before', ripa_final.shape[0])
ripa_final = ripa_final.drop_duplicates()
print('DF Shape After', ripa_final.shape[0])
ripa_final.shape
```
